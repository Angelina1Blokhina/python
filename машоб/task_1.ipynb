```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import statistics as st
import scipy.stats as stats
import pylab as pl
```

```python
X = [63, 57, 93, 55, 38, 53, 23, 0, 46, 51, 38, 103, 68, 60, 31, 55, 24, 50, 44, 97]
Y = [183, 257, 161, 149, 145, 208, 178, 193, 204, 136, 184, 157, 212, 127, 191, 187, 202, 220, 194, 196]
```

```python
srednee =np.mean(X)

mediana=np.median(X)

moda=st.mode(X)
print('Среднее значение ',srednee,'\n','Медиана ',mediana,'\n','Мода ',moda)
```

```python
Dy=np.var(Y)
print('Дисперсия', Dy)
```

```python
Mx = np.mean(X)
sigma = math.sqrt(np.var(X))
x = np.linspace(Mx - 3*sigma, Mx + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, Mx, sigma))
```

```python
#вероятность больше 58
x = sorted(X)  #sorted
sko = np.std(x)

norm_rv = stats.norm(Mx, sko)
fig, ax = plt.subplots()
x = np.linspace(norm_rv.ppf(0.0001), norm_rv.ppf(0.9999), 100)
ax.plot(x, norm_rv.pdf(x))
ax.fill_between(x[x>58], norm_rv.pdf(x[x>58]), np.zeros(len(x[x>58])))
p = 1 - norm_rv.cdf(58)

plt.show()
print('P(X>58) =',p)
```

```python
#Двумерное мат.ожидание и матрица ковариации

Mx=np.mean(X)
My=np.mean(Y)
Mxy=np.vstack((Mx, My))
print(Mxy)
COV=np.cov(X,Y)
print(COV)
```

```python
#корреляция
rXY=np.corrcoef(X,Y)[0,1]
print(rXY)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = X, y = Y)
plt.xlabel("Возраст")
plt.ylabel("Рост")

plt.show()
```

Большая вероятность достигается при 3

```python
x = np.linspace(-10,10,100)
xna=stats.norm.pdf(x,3,4)
xnb=stats.norm.pdf(x,2,3)

plt.plot(x, xna)
plt.plot(x, xnb)
idx = np.argwhere(np.diff(np.sign(xnb-xna))).flatten()
plt.plot(x[idx], xnb[idx], 'ro')

plt.show()
print(x[idx])
```
