```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def exponentialDistribution(busyTime):
    x=random.random()
    return(-math.log(x)*busyTime)
```

```python
def randomFunction():
    probability = random.randint (1, 100);
    if (probability<=4):
        return 1
    elif(probability>4 and probability<=12):
        return 2
```

```python
def mainAlgorithm(timeMachine1,timeMachine2,timeMachine3,timeMachine4,checkTimeMachine4,probabilityWasted,currentTime,
detailsMachine4,ﬁrstQueue,secondQueue,wasted,countDetails,detailInterval):
   
    ﬁrstMachineTime=40
    secondMachineTime=60
    thirdMachineTime=100
    fourthMachineTime=100
    lenMachine3_4=0
    while(countDetails!=500):
        ﬁrstQueue.append(0)
        #print(ﬁrstQueue)

        if(timeMachine1<=currentTime):
            newDetail=0
            while(ﬁrstQueue[newDetail]!=0):
                newDetail+=1
            timeMachine1=currentTime+exponentialDistribution(ﬁrstMachineTime)
            #detailsMachine1+=1
            probabilityWasted=randomFunction()
            if(probabilityWasted==1):
                ﬁrstQueue[newDetail]=1      
            else: 
                ﬁrstQueue.pop(newDetail)
                secondQueue+=1

        elif(timeMachine2<=currentTime and timeMachine1>currentTime):
            #detailsMachine2+=1
            if(ﬁrstQueue[0]==0):
                timeMachine2=currentTime+exponentialDistribution(secondMachineTime)
                probabilityWasted=randomFunction()
                if(probabilityWasted==2):
                     ﬁrstQueue[0]=1
                else:
                    ﬁrstQueue.pop(0)
                    secondQueue+=1

            elif(ﬁrstQueue[0]==1):
                timeMachine2=currentTime+exponentialDistribution(secondMachineTime)
                probabilityWasted=randomFunction()
                if(probabilityWasted==2):
                    wasted+=1
                    countDetails+=1
                    ﬁrstQueue.pop(0)
                else:
                    ﬁrstQueue.pop(0)
                    secondQueue+=1

        if(secondQueue<=3):
            if(timeMachine3<=currentTime):
                countDetails+=1
                secondQueue-=1
                timeMachine3=currentTime+exponentialDistribution(thirdMachineTime)
                #print("*")
        else:
            if(lenMachine3_4<secondQueue):
                lenMachine3_4=secondQueue
            if(timeMachine3<=currentTime):
                countDetails+=1
                secondQueue-=1
                timeMachine3=currentTime+exponentialDistribution(thirdMachineTime)
                
                #print("**")
            elif(timeMachine4<=currentTime and timeMachine3>currentTime):
                detailsMachine4+=1
                countDetails+=1
                secondQueue-=1
                checkTimeMachine4+=exponentialDistribution(fourthMachineTime)
                timeMachine4=currentTime+exponentialDistribution(fourthMachineTime)
                
                #print("***")

        currentTime+=exponentialDistribution(detailInterval)
      
    print('Время моделирования:',currentTime,"минут",'\nВремя работы 4го станка',checkTimeMachine4,"минут",
          '\nМаксимальный задел на вторичной обработке:' ,(lenMachine3_4/500)*100,"%",
          '\nЗабраковано деталей:', wasted,'  Процент брака:',(wasted/500)*100,'%','\nЗагрузка 4го станка:', (detailsMachine4/500)*100,'%')
    print(lenMachine3_4,detailsMachine4)
    return((lenMachine3_4/500)*100)
```

```python
timeMachine1=0
timeMachine2=0
timeMachine3=0
timeMachine4=0
checkTimeMachine4=0
probabilityWasted=0
currentTime=0
    

detailsMachine4=0
    
ﬁrstQueue=[]
secondQueue=0
    
wasted=0
countDetails=0
detailInterval=50    


m=mainAlgorithm(timeMachine1,timeMachine2,timeMachine3,timeMachine4,checkTimeMachine4,probabilityWasted,currentTime,
detailsMachine4,ﬁrstQueue,secondQueue,wasted,countDetails,detailInterval)
```

```python
z=[]

timeMachine1=0
timeMachine2=0
timeMachine3=0
timeMachine4=0
checkTimeMachine4=0
probabilityWasted=0
currentTime=0
    

detailsMachine4=0
    
ﬁrstQueue=[]
secondQueue=0
    
wasted=0
countDetails=0
#detailInterval=50    
ﬁrstMachineTime=40
secondMachineTime=60
thirdMachineTime=100
fourthMachineTime=100

for i in range(50,200):
    x=mainAlgorithm(timeMachine1,timeMachine2,timeMachine3,timeMachine4,checkTimeMachine4,probabilityWasted,currentTime,
detailsMachine4,ﬁrstQueue,secondQueue,wasted,countDetails,i)
    z.append(x)
    print(i,'минут ',x,'%')
x=np.arange(50, 200, 1)
y=z

fig, ax = plt.subplots()

ax.bar(x, y)
fig.set_figwidth(9)   
fig.set_figheight(6)    

plt.show()
```
