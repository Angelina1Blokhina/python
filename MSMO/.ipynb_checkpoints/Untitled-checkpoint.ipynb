```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
timeMachine1=0
timeMachine2=0
timeMachine3=0
timeMachine4=0

checkTimeMachine4=0

currentTime=0
#detailsMachine1=0
#detailsMachine2=0
detailsMachine3_4=0
detailsMachine4=0
ﬁrstQueue=[]
secondQueue=0

#wastedDetails=0 

wasted=0
countDetails=0

detailInterval=50
ﬁrstMachineTime=40
secondMachineTime=60
thirdMachineTime=100
fourthMachineTime=100

```

```python
def exponentialDistribution(busyTime):
    x=random.random()
    return(-math.log(x)*busyTime)
```

```python
def randomFunction():
    probability = random.randint (1, 100);
    if (probability<=4):
        return 1
    elif(probability>4 and probability<=12):
        return 2
```

```python
def mainAlgorithm(detailInterval):
    timeMachine1=0
    timeMachine2=0
    timeMachine3=0
    timeMachine4=0
    checkTimeMachine4=0
    probabilityWasted=0
    currentTime=0
    
    detailsMachine3_4=0
    detailsMachine4=0
    
    ﬁrstQueue=[]
    secondQueue=0
    
    wasted=0
    countDetails=0
    
    ﬁrstMachineTime=40
    secondMachineTime=60
    thirdMachineTime=100
    fourthMachineTime=100
    
    while(countDetails!=500):
        ﬁrstQueue.append(0)

        if(timeMachine1<=currentTime and ﬁrstQueue[0]==0):
            timeMachine1=currentTime+exponentialDistribution(ﬁrstMachineTime)
            #detailsMachine1+=1
            probabilityWasted=randomFunction()
            if(probabilityWasted==1):
                ﬁrstQueue[0]=1      
            else: 
                ﬁrstQueue.pop(0)
                secondQueue+=1

        elif(timeMachine2<=currentTime and timeMachine1>currentTime):
            #detailsMachine2+=1
            if(ﬁrstQueue[0]==0):
                timeMachine2=currentTime+exponentialDistribution(secondMachineTime)
                probabilityWasted=randomFunction()
                if(probabilityWasted==2):
                     ﬁrstQueue[0]=1
                else:
                    ﬁrstQueue.pop(0)
                    secondQueue+=1

            elif(ﬁrstQueue[0]==1):
                timeMachine2=currentTime+exponentialDistribution(secondMachineTime)
                probabilityWasted=randomFunction()
                if(probabilityWasted==2):
                    wasted+=1
                    countDetails+=1
                    ﬁrstQueue.pop(0)
                else:
                    ﬁrstQueue.pop(0)
                    secondQueue+=1

        if(secondQueue<=3):
            if(timeMachine3<=currentTime):
                countDetails+=1
                secondQueue-=1
                timeMachine3=currentTime+exponentialDistribution(thirdMachineTime)
                #print("*")
        else:
            if(timeMachine3<=currentTime):
                countDetails+=1
                secondQueue-=1
                timeMachine3=currentTime+exponentialDistribution(thirdMachineTime)
                detailsMachine3_4+=1
                #print("**")
            elif(timeMachine4<=currentTime and timeMachine3>currentTime):
                detailsMachine4+=1
                countDetails+=1
                secondQueue-=1
                checkTimeMachine4+=exponentialDistribution(fourthMachineTime)
                timeMachine4=currentTime+exponentialDistribution(fourthMachineTime)
                detailsMachine3_4+=1
                #print("***")

        currentTime+=exponentialDistribution(detailInterval)
    #print('',currentTime,'\n',checkTimeMachine4,'\n' ,detailsMachine3_4, '\n', wasted,'\n', detailsMachine4)
    return(detailsMachine3_4)
        
        
        
```
